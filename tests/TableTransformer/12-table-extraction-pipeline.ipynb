```python
import matplotlib.pyplot as plt
import networkx as nx
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(tr.G, scale=2, fixed=[0, 1])
nx.draw(tr.G, pos=pos, with_labels=True, node_size=100, font_size=0)
plt.show()
```

```python
# get the total number of nodes of type 'table'
tables = []
for node in tr.G.nodes(data=True):
    if node[1]['type'] == 'table':
        tables.append(node)
print(len(tables))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

plt.figure(figsize=(12, 12))

# Increase the value of 'k' to move nodes farther apart
# Remove or adjust the 'fixed' parameter as needed
pos = nx.spring_layout(tr.G, scale=5, k=0.15)

nx.draw(tr.G, pos=pos, with_labels=True, node_size=10, font_size=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

plt.figure(figsize=(12, 12))

# Increase the value of 'k' to move nodes farther apart
# Remove or adjust the 'fixed' parameter as needed
pos = nx.spring_layout(tr.G, scale=5, k=0.15, iterations=1)

nx.draw(tr.G, pos=pos, with_labels=True, node_size=10, font_size=0)
plt.show()
```
